# Summary: Speed

This file provides concise summary to the speed benchmarks provided previously,

1. Fifty tests were performed with varying number of x-values in the x-array to be evaluated. 
2. In the majority of test cases, the `pymultifit` implementation performs faster then `scipy`.
3. However, distributions such as `beta`, `foldnorm` and `halfnorm` consistently show performance parity with `scipy`. Notable performance degradations are observed in these distributions. 
4. Other distributions, such as `skewnorm` and `foldnorm`, exhibit some variability, with occasional high computation time from `pymultifit`.

Overall, the benchmarks suggest that `pymultifit` implementations generally offers better performance over a large values of x-array sizes.
