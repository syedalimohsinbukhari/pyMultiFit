# 01. Generate ``Gaussian`` data

This first tutorial is aimed at the very basic statistical use of the ``pyMultiFit`` library, that is to generate a simple ``Gaussian`` data with $\mu = 0$ and $\sigma = 1$.

```python
import matplotlib.pyplot as plt
import numpy as np
from pymultifit.distributions import GaussianDistribution
```

```python
gdist = GaussianDistribution(normalize=False)
```

```python
x = np.linspace(-5, 5, 10_000)
y = gdist.pdf(x)
```

```python
plt.plot(x, y)
plt.show()
```
