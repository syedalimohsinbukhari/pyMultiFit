# 01. Generate Gaussian data

This first tutorial is aimed at the very basic use of the ``pyMultiFit`` library, that is, to generate Gaussian data with $\mu = 0$ and $\sigma = 1$. This can be done via the **GaussianDistribution** class, and for comparison, we will also import the ``scipy.stats.norm`` function.

```python
from pymultifit.distributions import GaussianDistribution
from scipy.stats import norm
```

Calling the Gaussian distribution is as easy as writing **GaussianDistribution**.

```python
gdist1 = GaussianDistribution()
gdist2 = norm()
```

Next, we set up our x array to generate the gaussian on using `numpy`,

```python
import numpy as np

x = np.linspace(-5, 5, 10_000)
y1 = gdist1.pdf(x)
y2 = gdist2.pdf(x)
```

Finally, let's see the plots for both of these distributions,

```python
import matplotlib.pyplot as plt

plt.plot(x, y1, label='pyMultiFit \nGaussian Distribution')
plt.plot(x, y2, label='scipy \nGaussian Distribution')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Gaussian Distribution')
plt.legend()
plt.tight_layout()
plt.show()
```

But what's this? The distributions don't align up. This is intended, the distributions created by the ``pyMultiFit`` library is intended to be **un-normalized**, more on this later. In order to generate a **normalized** distribution, we can set the `normalize` argument to `True`.

```python
gdist1 = GaussianDistribution(normalize=True)
y1 = gdist1.pdf(x)
```

Now, if we re-plot both the distributions, we get,

```python
plt.plot(x, y1, label='pyMultiFit \nGaussian Distribution')
plt.plot(x, y2, label='scipy \nGaussian Distribution')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Gaussian Distribution')
plt.legend()
plt.tight_layout()
plt.show()
```

Which is exactly what we wanted.
